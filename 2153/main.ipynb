```python
import pandas as pd
```

```python
class Bus:
    def __init__(self,_id,arrival_time,cap):
        self.id=_id
        self.time=arrival_time
        self.capacity=cap
        self.passengers=[]
        self.is_full=False
    def add(self,time,pass_id):
        if self.is_full==True:
            return False
        if self.capacity>0:
            if self.time>time:
                self.capacity-=1
                self.passengers.append(pass_id)
                if self.capacity==0:
                    self.is_full=True
                return True
            return False
        return False
```

```python
buses={}
_input=pd.read_csv("buses.csv")
for i in range(len(_input)):
    row=_input.iloc[i]
    if row.bus_id not in buses:
        b=Bus(row.bus_id,row.arrival_time,row.capacity)
        buses[row.bus_id]=b
b=list(buses.values())
b.sort(key=lambda x: x.time, reverse=True)
# for i in b:
#     print (i.time,i.capacity)
```

```python
def add_pass_to_bus(pass_id,time):
    flag=False
    
    for bus in b:
        if flag==False:
            temp=bus.add(time,pass_id)
            if temp==True:
                flag=True
        
```

```python
%%time
passengers=pd.read_csv("passengers.csv")
for i in range(len(passengers)):
    row=passengers.iloc[i]
    add_pass_to_bus(row.passenger_id,row.arrival_time)
for bus in b:
    print(bus.time,bus.capacity,bus.passengers)
```
