```python
def gen_date():
    years=range(2022,2023)
    monthes=range(1,13)
    days=range(1,32)
    return str(choice(years))+"-"+str(choice(monthes))+"-"+str(choice(days))
```

```python

trips_cols=['id-int','client_id-int','driver_id-int','city_id-int','status-enum','request_date-date']
class Trip:
    def __init__(self,_id,client_id,driver_id,city_id,status,request_date):
        self.id=_id
        self.client_id=client_id
        self.driver_id=driver_id
        self.city_id=city_id
        self.status=status
        self.date=request_date
    def toString(self):
        return str(self.id)+","+str(self.driver_id)+","+str(self.status)+","+str(self.date)
```

```python
user_cols=["user_id-int","banned-enum","role-enum"]
class User:
    def __init__(self,user_id,banned,role):
        self.user_id=user_id
        self.banned=banned
        self.role=role
    def toString(self):
        return str(self.user_id)+","+str(self.banned)+","+str(self.role)
```

```python
from random import randint,choice
def gen_data(num_of_rows):
    res={
        "trips":[],
        "clients":[],
        "drivers":[]
    }
    b=["yes","no","no","no","no","no"]
    list_of_clients=range(num_of_rows)
    list_of_drivers=range(num_of_rows)
    cities=range(1,20)
    role=["client","driver"]
    statuses=["completed","completed","completed","completed","cancelled_by_driver","cancelled_by_client"]
    
    for i in range(num_of_rows):
        temp_user=User(i,choice(b),choice(role))
        if temp_user.role=="client":
            res["clients"].append(temp_user)
        else:
            res["drivers"].append(temp_user)
        
    for i in range(int(num_of_rows/2)):
        temp_trip=Trip(i,choice(res["clients"]).user_id,
                       choice(res["drivers"]).user_id,choice(cities),
                      choice(statuses),gen_date())
        res["trips"].append(temp_trip)
                    
    
    return res
```

```python
import time
t1=time.time()
data=gen_data(10000)

t2=time.time()
print(t2-t1)

```

```python
users_dict={}
dates_dict={}
for u in data["clients"]:
    users_dict[u.user_id]=u
for d in data["drivers"]:
    users_dict[d.user_id]=d
for d in data["trips"]:
    dates_dict[d.date]={
        "canceled":0,
        "completed":0
    }
for t in data["trips"]:
    if t.status=="completed":
        dates_dict[t.date]["completed"]+=1
    else:
        if users_dict[t.client_id].banned=="no" and users_dict[t.driver_id].banned=="no":
            dates_dict[t.date]["canceled"]+=1
            

for i in dates_dict:
    print(i,dates_dict[i])

```
