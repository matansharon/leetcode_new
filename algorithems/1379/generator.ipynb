```python

from numpy import random
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
root=TreeNode(0)
#generate a random tree
for i in range(10):
    cnt=random.randint(0, 100)
    if i == 0:
        root = TreeNode(cnt)
    else:
        cur = root
        while True:
            if cnt < cur.val:
                if cur.left == None:
                    cur.left = TreeNode(cnt)
                    break
                else:
                    cur = cur.left
            else:
                if cur.right == None:
                    cur.right = TreeNode(cnt)
                    break
                else:
                    cur = cur.right
#inorder traversal
def inorder(root):
    if root == None:
        return
    inorder(root.left)
    print(root.val)
    inorder(root.right)
inorder(root)
#preorder traversal
def preorder(root):
    if root == None:
        return
    print(root.val)
    preorder(root.left)
    preorder(root.right)
preorder(root)
#postorder traversal
def postorder(root):
    if root == None:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.val)
postorder(root)
#level order traversal
def levelorder(root):
    if root == None:
        return
    q = [root]
    while len(q) > 0:
        cur = q.pop(0)
        print(cur.val)
        if cur.left != None:
            q.append(cur.left)
        if cur.right != None:
            q.append(cur.right)
levelorder(root)

    
```

```python
class Solution:
    def getTargetCopy(self, original: TreeNode, cloned: TreeNode, target: TreeNode) -> TreeNode:
        while True:
            if original == target:
                return cloned
            if original.left != None:
                res = self.getTargetCopy(original.left, cloned.left, target)
                if res != None:
                    return res
            if original.right != None:
                res = self.getTargetCopy(original.right, cloned.right, target)
                if res != None:
                    return res
            return None
s = Solution()

res = s.getTargetCopy(root, root, root.right)
root.right.val, res.val
```
